```python
# AUTHOR: Hadeel S

def create_acronyms(data):
    acs2 = []
    longf = []

    acros = data.acronyms.values
    txt = data.text.values
    zipped = list(zip(txt,acros))
    for z in zipped:
        acs = []
        for i in range(len(z[1])):
            acs.append(z[0][z[1][i][0]:z[1][i][1]])
        acs2.append(acs)
    data['acronyms-text'] =acs2
    
# Creating long forms from here

    longform = data["long-forms"].values
    zipped2 = list(zip(txt,longform))
    for z2 in zipped2:
        longform_temp = []
        for i in range(len(z2[1])):
            longform_temp.append(z2[0][z2[1][i][0]:z2[1][i][1]])
        longf.append(longform_temp)
    data["long-forms-text"]= longf
    
    return data
```

```python
englishDFTrainScientificResults = create_acronyms(englishDFTrainScientificResults)
englishDFDevScientificResults = create_acronyms(englishDFDevScientificResults)
englishDFTrainLegalResults = create_acronyms(englishDFTrainLegalResults)
englishDFDevLegalResults = create_acronyms(englishDFDevLegalResults)
danishDFTrainResults = create_acronyms(danishDFTrainResults)
danishDFDevResults = create_acronyms(danishDFDevResults)
frenchDFTrainResults = create_acronyms(frenchDFTrainResults)
frenchDFDevResults = create_acronyms(frenchDFDevResults)
persianDFTrainResults = create_acronyms(persianDFTrainResults)
persianDFDevResults = create_acronyms(persianDFDevResults)
spanishDFTrainResults = create_acronyms(spanishDFTrainResults)
spanishDFDevResults = create_acronyms(spanishDFDevResults)
vietnameseDFTrainResults = create_acronyms(vietnameseDFTrainResults)
vietnameseDFDevResults = create_acronyms(vietnameseDFDevResults)
```
